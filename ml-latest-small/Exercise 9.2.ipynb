# Building a recommender system using Movie lens Data set

```python
#Load the data from the csv files

import numpy as np
import pandas as pd
import warnings

ratings_df = pd.read_csv('ratings.csv')
movies_df = pd.read_csv("movies.csv")
ratings_df.head(10)
```

```python
movies_df.head(10)
```

We need to merge the datsets to bring two datasets together into one.

```python
# Merge the two data sets
ratings_df = ratings_df.merge(movies_df,on='movieId', how='left')
warnings.filterwarnings('ignore')
ratings_df.head(10)
```

The dataset we have is a collection of ratings by a number of users for different movies. We need to find out the average rating for each and every movie in the dataset.

```python
# Calculate the average ratings of the movies
Average_ratings_df = pd.DataFrame(ratings_df.groupby('title')['rating'].mean())
Average_ratings_df.head(10)
```

No. of ratings for a movie also plays a crucial role in the overall rating of the movie. Therefore, we will also consider the total ratings cast for each movie.

```python
# Calculate the total ratings of the movies
Average_ratings_df['Total Ratings'] = pd.DataFrame(ratings_df.groupby('title')['rating'].count())
Average_ratings_df.head(10)
```

We will create a pivot table where the rows are userIds and the columns represent the movies. The values of the matrix represent the rating for each movie by each user.

```python
# Pivot the data to show the rating for each movie by each user.
movie_user_df = ratings_df.pivot_table(index='userId',columns='title',values='rating')
movie_user_df.head()
```

To find the correlation value for the movie with all other movies in the data, we will pass all the ratings of the picked movie to the corrwith method to compute the pairwise correlation.

```python
# Take the input from user and calculate the correlation with other movies
movie_name=input("enter the movie name:")
correlations = movie_user_df.corrwith(movie_user_df[movie_name])
correlations.head()
```

Join the correlations that we got with the total ratings to be able to filter out the ones with no ratings and to have the ability to filter on no. of ratings

```python
# For filtering out the recommendations, drop all the blank rows
recommendation = pd.DataFrame(correlations,columns=['Correlation'])
recommendation.dropna(inplace=True)
recommendation = recommendation.join(Average_ratings_df['Total Ratings'])
recommendation.head()
```

```python
# Filter out recommendations with more than 75 ratings and display top 10
recc = recommendation[recommendation['Total Ratings']>75].sort_values('Correlation',ascending=False).reset_index()
recc = recc.merge(movies_df,on='title', how='left')
recc.head(10)
```
